```python
import pandas as pd
import os
from mapboxgl.utils import *
from mapboxgl.viz import *
```

```python
df = pd.read_csv("~/Dropbox/locus/data/jakarta_combined/x.txt",sep="|").round(4)
df.head()
```

```python
token = 'MAPBOX_ACCESS_TOKEN'
```

```python
# Create a geojson file export from the current dataframe
df_to_geojson(df, 
              filename='../data/combined.geojson',
              properties=['name','street','number'], 
              lat='lat', 
              lon='lng', 
              precision=3)
```

```python
# Just show a map of the data
viz = CircleViz('data/combined.geojson', 
                access_token=token, 
                radius=2, 
                color_default='yellow',
                center=(106.8,-6.2), 
                zoom=10)

'''
viz = LinestringViz('data/jakarta_timur.geojson',
                    access_token=token,
                    color_property='name',
                    data_join_property='name',
                    line_stroke='--',
                    line_width_default=2,
                    opacity=1.0,
                    center=(106.8,-6.2),
                    zoom=10)
'''

# Map settings
viz.style = 'mapbox://styles/mapbox/dark-v10?optimize=true'
viz.label_color = 'hsl(62, 100%, 50%)'
viz.label_halo_color = 'hsla(0, 0%, 10%, 0.75)'

viz.show()
```
